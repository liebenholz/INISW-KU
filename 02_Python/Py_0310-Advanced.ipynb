중첩을 하는 이유는 무엇인가?

Closure: 두 개 이상의 함수를 활용해 다양한 방법으로 응용하고 재사용할 수 있다.

유연하게 설계를 하기 때문에, 구조체를 만들 때 쉽게 관리할 수 있다는 장점이 있다.

```python
class X:
  def __init__(self, x):
    self.x = x
  def __call__(self, y):
    return self.x + y
```

```python
X(3)(4) # 동작하는 방법은 함수와 같지만, 접근 가능 여부에 차이가 있다.
```

```python
# tenseorflow
class X:
  def __init__(self, x):
    self.x = x

  def call(self, y):
    return self.x + y

  def __call__(self, y):
    self.call(y)

# pytorch는 def forward(self, x):
```

**책임 분리의 원칙**이라고도 부른다.

잘게 쪼개놓음으로써 필요한 부분만 바꿀 수 있는 **유연성** 측면에서 이점이 있다.

```python
class A:
  # 값을 만들어준다
  def __new__(cls):
    print('new')
    return super().__new__(cls)

  # 만들어준 값을 초기화한다.
  def __init__(self):
    print('init')
    # return 1
```

```python
A() # 괄호를 붙였을 때 new가 먼저 실행된 후, init가 실행됨. 책임 분리의 원칙.
```
