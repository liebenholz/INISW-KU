### with 구문

```python
class A:
  # with를 사용하려면 아래 두 개가 정의되어있어야 한다.
  # duck typing으로 구현
  def __enter__(self):
    print('enter')
  def __exit__(self, exc_type, exc_val, exc_tb):
    print('exit')
```

```python
with A():
  # enter(open) -> with -> exit(하자마자 close)
  print('with')
```

```python
class B:
  pass
```

```python
# 위는 아래의 중첩구문
with A() as a, B() as b:
  pass

with A() as a:
  with B() as B:
    pass
```

# Mastering Python Decorators

```python
def x(fun):
  def xx(*a, **b): # 모든 입력 대응 가능, 인자 맞추기 위해서
    print('d')
    fun(*a, **b)
  return xx
```

```python
@x
def t():
  print('t')
```

```python
t # <function __main__.x.<locals>.xx(*a, **b)>
```

```python
print(t)
```

```python
t()
```

```python
@x
def tt(a):
  print('t')
```

```python
tt(1)
```

```python
from functools import wraps

def x(fun):
  @wraps(fun)
  def xx(*a, **b):
    print('d')
    fun(*a, **b)
  return xx
```

```python
@x
def t():
  print('t')
```

```python
t # <function __main__.x.<locals>.xx(*a, **b)>
```

### Singleton

면접에 많이 나온다... GoF 디자인 패턴에 대해 공부할 것.

```python
class Singleton(type):
    instance = None
    def __call__(cls, *args, **kw):
      print('aa')
      if not cls.instance: # instance가 없으면 새로 만들고
          cls.instance = super(Singleton, cls).__call__(*args, **kw)
      return cls.instance  # 있으면 기존 것을 만드세요
```

```python
class AS(metaclass=Singleton):
  pass
```

```python
a = AS() # 새로 만들었음
```

```python
a.x = 1
b = AS()
```

```python
b.x # 기존 것을 가져왔음
```

```python
a is b
```
