클래스는 함수와 같은 역할을 하지만, 디테일은 다르다. 클래스가 좀 더 유연함.

```python
class A:
  a = 1         # a: class 변수, class가 직접 사용할 수 있음
  def x(self):  # x: (instance) method,
    self.b = 1  # b: instance 변수, instance method가 실행될 시점에 생성
    print('x')

  @classmethod
  def y(cls):   # classmethod, class가 직접 사용할 수 있음
    cls.c = 1
    print('y')

  @staticmethod # staticmethod, class가
  def z():
    print('z')
```

## Class & Instance

```python
A.a # class가 직접 사용할 수 있음
```

```python
A.b # instance 변수기 때문에 class가 직접 사용 불가
```

```python
# 클래스 기반으로 생성된 인스턴스를 이용한 프로그래밍이 OOP
aa = A() # aa: object, instance # a = int(1)
```

```python
dir(aa)
```

```python
aa.b # b가 정의되어 있지 않기 때문에 AttributeError
```

```python
aa.x() # 인스턴스 실행
```

```python
aa.b # 인스턴스 실행 후 생성됨
```

```python
# 실수를 줄이기 위해서는 __init__에 인스턴스 변수를 선언할 것
class A():
  def __init__(self):
    self.b = 1
```

```python
aaa = A()
vars(aaa) # 현재 정의되어 있는 인스턴스 변수를 나타냄, 동적으로 수정 가능.
```

```python
class B:
  bb = 1
```

```python
b = B()
b.t = 1 # 동적으로 인스턴스 변수 추가
vars(b)
```

```python
c = B()
```

```python
vars(c)
```

## Static Method

```python
# 인스턴스화 후 사용해야 하는 문제점이 발생
class Math: # 클래스 이름이 namespace
  def sin(self,x):
    print('sin')
  def cos(self,x):
    print('cos')

  @staticmethod  # 바로 접근할 수 있음, 유지보수 간편
  def tan(x):
    print('tan')
```

```python
a = Math()
a.sin(3) # 불편함... 함수처럼 쓸 수 있는 방법은 없을까?
```

```python
Math.tan(3)
```

```python
a.tan(3) # 함수처럼 직접 접근할 수도, 인스턴스로 접근할 수도 있고, 하나의 구성요소로도 사용할 수 있음(유연함)
```
