### 다중 상속

두 명 이상의 부모를 가질 수 있다. 대신, 어느 누구한테 어떤 기능을 상속받았는지 알 수 없다.

코드를 복잡하게 만들기 때문에 가능하면 하나만을 상속받는 것이 바람직하다(다이아몬드 문제)

파이썬에서는 **MRO**를 지원하여 다중 상속에서 다이아몬드 문제를 해결했다. : Method Resolve Order

```python
class AA:
  x = 1

class BB:
  x = 2
```

```python
DD.mro() # 메타클래스 안에 있는 기능
```

```python
class EE(AA,DD): # 순서가 있으니 주의할 것
  pass
```

함수는 파라미터의 개수에 맞춰서 실행해야 한다(default 제외)

메소드를 사용할 때는 첫 번째 인자는 생략할 수 있지만, 함수로 사용할 때는 생략할 수 없다(self)

self는 instance 그 자체

```python
# 다중상속의 경우, 순서가 중요하다.

class A:
  def __init__(self):
    print('A')

class B(A):
  def __init__(self):
    A.__init__(self) # 함수임
    print('B')

class C(A):
  def __init__(self):
    A.__init__(self) # 함수임
    print('C')

class D(B, C):
  def __init__(self):
    B.__init__(self) # 함수임
    C.__init__(self) # 함수임
    print('D')
```

```python
b = B()
```

```python
d = D() # 다중상속 시 부모의 영향으로 여러 번 실행될 수 있다.
```

```python
# 다중상속의 경우, 순서가 중요하다.

class A:
  def __init__(self):
    print('A')

class B(A):
  def __init__(self):
#     A.__init__(self) # 함수임
    super(B, self).__init__() # B의 부모를 instance화한다.
    print('B')

class C(A):
  def __init__(self):
#     A.__init__(self) # 함수임
    super().__init__() # 파이썬3부터 self를 생략할 수 있다.
    print('C')

class D(B, C):
  def __init__(self):
#     B.__init__(self) # 함수임
#     C.__init__(self) # 함수임
    super().__init__() # 모든 부모를 탐색해서 사용할 수 있다.
    print('D')
```

```python
d = D() # A 실행을 한 번만, 다중상속 체계를 고려해서 MRO에 따라 부모를 실행시켜준다.
```

```python
D.mro() # 스택 영역에 넣기 때문에 역순으로 나온다.
```

## Reduce / Accumulation

Reduce: 통계의 기본, 여러 개의 결과값을 하나로 표현한다.

Accumulation: 누적

```python
from functools import reduce
```

```python
reduce(lambda x,y:x+y, [0,1,2,3]) # 하나의 결과값으로 차원을 줄여준다
```

```python
from itertools import accumulate
```

```python
t = accumulate([0,1,2,3], lambda x,y:x+y)
```

```python
next(t) # 0, 1, 3, 6 순으로 출력
```

```python
list(filter(lambda x:x>1 or None, [1,2,3]))
```
