```python
resp = urlopen('https://www.naver.com')
msg = resp.read()
```

```python
# scheme => https://
# host   => www.google.com
# path   => /search?
# params => q: %EC%B9%B4%EB%A6%AC%EB%82%98   ===> Percent Encoding %(Hexadecimal 16진수)
#           sourceid: chrome
#           ie: UTF-8

#          0000 0000
#          0111 1111 => ASCII
#          1~ => Non ASCII
# 2 bytes => 조합형, 완성형, 확장완성형
# UNICODE
# ASCII => 1byte 65:A, 97:a
# (str)A-65 => (bytes)A-65
# Non ASCII => 1~ 0~(65) => A? => Charser(Encoding)
# (str) => (Bytes) 16진수 => Percent
# HTML => Markup => <Tag></Tag>
# 1 < 3, 기호들 => HTML Entities
# < => &; => &lt; 한글=>10진수 &10진수;
```

```python
from urllib.parse import quote, unquote
```

```python
quote('카'), unquote('%EB%82%98')
```

```python
url = 'https://www.google.com/search?q=카리나&sourceid=chrome&ie=UTF-8'
url = 'https://www.google.com/search?q='+quote('카리나')+'&sourceid=chrome&ie=UTF-8'
resp = urlopen(url)
```

```python
# Non ASCII -> quote -> Bytes(Percent Encoding-Hexadecimal) 형태로 변환해야 함
unquote(resp.geturl())
```

```python
# resp.getheaders() => content-type: ISO-8859-1
# resp.read().decode()
from html import unescape, escape
src = '&#47751; &#52488; &#50504;&#50640; &#51060;&#46041;&#54616;&#51648; &#50506;&#45716; &#44221;&#50864;'
unescape(src)
```
