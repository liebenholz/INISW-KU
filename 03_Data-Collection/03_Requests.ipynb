```python
# HTTP - urllib - requests
# Client ----------- HTTP ------------ Server
#          Request          Response   => TCP/IP
#           Header            Body     => Bytes
#
# Opt-out(Blacklist -> 거부를 명시적으로 표현) => 기록을 남겨야 함
# 1. Robots.txt => Bot(Agent) Enable, Disable 명식적으로 표현됨
#    ; urllib.robotparser 잘 안 돌아감
#    ; HTML-HEAD-META noindex, nofollow 확인
# 2. Traffic => 과도한 요철 X => delay
# 3. 개인정보 => 어떤 경우에도 X
# 4. 저작권 => DB에 있는 자료 X
# 5. 약관 확인 => 현실적으로 어려움
# 6. 안전한 방법 => API
# 공부 및 연구용, 내 데이터 내가 직접 수집, 배포X
```

```python
from urllib.error import HTTPError

url = 'https://httpbin.org/status/403'
try:
  urlopen(url)
except HTTPError as e:
  # if e.status != 200
  if 400 <= e.status < 500:
    print(e.status)
    print(e.reason)
    print(e.headers)
  elif 500 <= e.status:
    print('Try Again!')
```

## Requests

```python
from requests import request, get, post

url = 'https://httpbin.org/get'
resp = request('GET', url=url, params=params)
```

```python
params # 파이썬에서 사용하는 객체를 그대로, urllib에서는 string으로 풀어야 함
```

```python
#.                  urllib          requests
# 1. params     urlencode(->str)    dict O
# 2. request       Request 생성      자동
# 3. response   (임시)HTTP.response  request.response
# 4. 추가작업      read, decode       read X, bytes str 모두 제공, response.body를 객체로 들고 있음
# 5. Exception   HTTPError Throw    raise_for_status(), status_code 확인
# 6. JSON        json package 이용   포함, 더 상위 레벨이 있음
```
