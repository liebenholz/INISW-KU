파이썬의 From... 이전에 import에 대해

import: 기존에 작업해 둔 것을 가져오는 것.
- import sun

이름이 같으면 뒤엎는다. 이름의 충돌을 막기 위해 별칭을 사용할 수 있다.
- import sun as s

파일을 찾는 규칙
- sys.path 순서대로 찾는다. 여러 개가 있으면 가장 앞에 있는 것 가져온다.



````python
help('from')
````

다음을 `sun.py`로 저장한 후 가져오는 방법

```python
a = 1
_a = 2

def x():
  print("x")
  
def _x():
  print("_x")
```

다음과 같이 sun과 sun 내부의 함수 x를 가져올 수 있다.

```python
import sun
import sun as s
from sun import x # 내부의 일부 함수만 가져올 수 있다.
```

'*'는 모든 것을 뜻한다.
 > 삶, 우주, 그리고 모든 것에 대한 궁극적인 질문에 대한 해답. = 42(0101010)
 - 은하수를 여행하는 히치하이커를 위한 안내서, 더글러스 애덤스

```python
from sun import * # *는 전체
```

```python
__all__ = ['x']
```
로 지정해 *를 import하면 x만 import 가능하다.
