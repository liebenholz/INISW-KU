## User item Matrix

### Stack/Unstack
[Stack은 column을 index로(SCI), Unstack은 index를 column으로](https://pandas.pydata.org/docs/user_guide/reshaping.html
)

```python
data2.set_index(['user_id', 'item_id'])
```

```python
# stack/unstack(가장 어려움), index 뽑아오기, index를 column으로 바꾸기
# + value 내 missing data 채우기
data3 = data2.set_index(['user_id', 'item_id']).unstack().fillna(0)
data3
```

Colaborative Algorithm: 느리다

```python
data3.index
```

```python
# data3.values
data3.to_numpy()
```

```python
data4 = data3.T
data4
```

## Similarity

```python
data5 = data4.corr()
data5
```

```python
data5[3].sort_values(ascending=False)
```

```python
data5[3].nlargest(3)
```

```python
data5[3].nsmallest(3)
```

```python
def sim_me(user_id, n=5):
  return data5[user_id].nlargest(n+1).index[1:]
```

```python
sim_me(423)
```

## Find Relatives
나와 유사한 사람들이 했지만, 내가 경험하지 않은 것 찾기(차집합 B-A)

```python
# set(data[data['user_id'].isin(sim_me(1))]['item_id'].values)
```

```python
# set(data[data['user_id']==1]['item_id'].values)
```

```python
def not_shown_movie_id(user_id):
  # (나랑 유사한 사람들이 본 영화) - (내가 본 영화)
  return set(data[data['user_id'].isin(sim_me(user_id))]['item_id'].values) - set(data[data['user_id']==user_id]['item_id'].values)
```

```python
not_shown_movie_id(3)
```

```python
def not_shown_movie_title(user_id):
  return item[item['movie_id'].isin(not_shown_movie_id(user_id))]['movie_title'].values
```

```python
not_shown_movie_title(3)
```
