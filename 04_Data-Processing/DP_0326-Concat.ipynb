## Concatenate

```python
import numpy as np

a = np.arange(6)
a
```

```python
np.concatenate([a, a])
```

```python
aa = np.arange(6).reshape(2,3)
aa
```

```python
np.concatenate([aa, aa])
```

```python
np.concatenate([aa, aa], axis=1) # reduction 없음, 2차원 데이터일 때 다르게 작동한다.
```

```python
import pandas as pd
import seaborn as sns

tips = sns.load_dataset('tips')

tips0 = tips.iloc[0:4]
tips1 = tips.iloc[4:10]

tips0
```

```python
pd.concat([tips0, tips1])
```

```python
pd.concat([tips0, tips1], axis=1)
```

```python
tips_1 = tips.iloc[1]
tips_2 = tips.iloc[2]

tips_1
```

```python
pd.concat([tips_1, tips_2], axis=1)
```

## append

```python
np.append(a, 3)
```

```python
np.append(a, a) == np.concatenate((a, a))
```

## Stack
넘파이와 판다스의 스택은 다르다.

```python
np.hstack((aa, aa)) # Horizonal
```

```python
np.vstack((aa, aa)) # Vertical
```

```python
np.dstack((aa, aa)) # Depth, 차원 증가
```

```python
aa.shape
```

```python
np.dstack((aa, aa)).shape # z축으로 차원 확장
```

```python
np.dstack((aa, aa, aa)).shape
```

```python
np.stack((aa, aa), axis=0) # 차원을 하나 더 높여줌
```

```python
np.stack((aa, aa), axis=0).shape
```

```python
np.stack((aa, aa), axis=1)
```

```python
np.stack((aa, aa), axis=1).shape
```

```python
np.stack((aa, aa), axis=2)
```

```python
np.stack((aa, aa), axis=2).shape
```

```python
np.column_stack((aa,aa)) # == hstack()
```

```python
np.row_stack((aa,aa)) # == vstack()
```

```python
np.c_[aa, aa]
```

```python
np.r_[aa, aa]
```
