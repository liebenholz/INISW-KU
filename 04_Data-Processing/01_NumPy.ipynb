## Homogeneous

똑같은 데이터 타입의 집합

```python
import numpy as np # GPU 지원 안함
import array
```

```python
a = array.array('i', [1, 2, 3, 4])
```

```python
a # 빠르다
```

```python
a + a # 시퀀스 데이터 타입이라 수학적 컨셉이 없음
# Operator Overloading 사용해야 함
```

```python
[1,2,3] + [4,5,6]
```

숫자를 다루기 위한 파이썬 라이브러리, 수학적 컨셉, 벡터화

```python
def test_sum(a,b):
  return a + b
```

```python
test_sum([1,2,3],[4,5,6])
```

```python
@np.vectorize
def test_sum(a,b):
  return a + b
```

```python
test_sum([1,2,3],[4,5,6])
```
